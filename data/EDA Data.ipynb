```python
import sqlite3
import pandas as pd

# Path to your SQLite database file
db_file = r"C:/Users/I745988/Downloads/2023-10-25_cti_data_majd/2023-10-25_cti_data_majd.db"

# Connect to the SQLite database
conn = sqlite3.connect(db_file)

# Function to load table data into a DataFrame
def load_table_to_dataframe(conn, table_name):
    query = f"SELECT * FROM {table_name}"
    return pd.read_sql_query(query, conn)

# List of tables to load
tables_to_load = ['pulses', 'indicators', 'ip_location']

# Load data into DataFrames
dataframes = {table: load_table_to_dataframe(conn, table) for table in tables_to_load}

# Access individual DataFrames
pulses_df = dataframes['pulses']
indicators_df = dataframes['indicators']
ip_location_df = dataframes['ip_location']

# Example: Display the first few rows of each DataFrame
print("Pulses DataFrame:")
print(pulses_df.head())

print("\nIndicators DataFrame:")
print(indicators_df.head())

print("\nIP Location DataFrame:")
print(ip_location_df.head())

# Close the database connection
conn.close()
```

```python
# Function to compute value counts for all columns in a DataFrame
def display_value_counts(df, df_name):
    print(f"Value counts for DataFrame: {df_name}")
    for column in df.columns:
        print(f"\nColumn: {column}")
        print(df[column].value_counts())
    print("\n" + "-"*50 + "\n")

# Display value counts for each DataFrame
display_value_counts(pulses_df, "Pulses")
display_value_counts(indicators_df, "Indicators")
display_value_counts(ip_location_df, "IP Location")
```

```python
import random

# Add a new column 'ip' to indicators_df
# Randomly sample IPs from the 'ip' column of ip_location_df
indicators_df['ip'] = [random.choice(ip_location_df['ip'].dropna().tolist()) for _ in range(len(indicators_df))]

# Verify the changes
print("Updated indicators DataFrame with random IPs:")
print(indicators_df.head())
```

```python
import pandas as pd

# Assuming you have two DataFrames: indicators_df and ip_locations_df

# Perform the join using 'ip' as the key
merged_df = pd.merge(indicators_df, ip_location_df, on='ip', how='inner')  # Use 'inner' for inner join, you can adjust as needed

# Display the merged DataFrame
print(merged_df)
```

```python
# Assuming merged_df from previous merge and pulses_df are already defined

# Perform the merge using 'pulse_id' from merged_df and 'id' from pulses_df
final_merged_df = pd.merge(merged_df, pulses_df, left_on='pulse_id', right_on='id', how='inner')

# Display the final merged DataFrame
print(final_merged_df)
```

```python
final_merged_df.to_csv("final_data.csv", index=False)
```

```python
# Function to convert columns to datetime in a DataFrame
def convert_to_datetime(df, columns):
    for col in columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
    return df

# Specify datetime columns for each DataFrame
datetime_columns_pulses = ['created_at', 'updated_at']  # Replace with actual datetime columns in pulses_df
datetime_columns_indicators = ['indicator_created', 'indicator_expiration']  # Replace with actual datetime columns in indicators_df
datetime_columns_ip_location = ['last_seen']  # Replace with actual datetime columns in ip_location_df

# Convert datetime columns in each DataFrame
pulses_df = convert_to_datetime(pulses_df, datetime_columns_pulses)
indicators_df = convert_to_datetime(indicators_df, datetime_columns_indicators)
ip_location_df = convert_to_datetime(ip_location_df, datetime_columns_ip_location)

# Verify the conversion
print("Datetime conversion completed:")
print("\nPulses DataFrame:")
print(pulses_df[datetime_columns_pulses].info())

print("\nIndicators DataFrame:")
print(indicators_df[datetime_columns_indicators].info())

print("\nIP Location DataFrame:")
print(ip_location_df[datetime_columns_ip_location].info())
```
